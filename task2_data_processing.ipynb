```python
import pandas as pd
import numpy as np
```

Task 1 — Load the JSON File

```python
#Converting the json file Data folder to a dataframe using the pandas .read_json() method
df = pd.read_json("data/trends_20260414.json")
print(f"Loaded {len(df)} rows")
```

Task 2 — Clean the Data

```python
df.head(30)
```

```python
# 1. Remove duplicates based on post_id
df = df.drop_duplicates(subset="post_id")
print(f"After removing duplicates: {len(df)}")
```

```python
# 2. Remove rows with missing critical values
df = df.dropna(subset=["post_id", "title", "score"])
print(f"After removing nulls: {len(df)}")
```

```python
# 3. Fix data types
# Convert score and num_comments to integers
df["score"] = pd.to_numeric(df["score"]).fillna(0).astype(int)
df["num_comments"] = pd.to_numeric(df["num_comments"]).fillna(0).astype(int)
```

```python
# 4. Remove low quality stories (score < 5)
df = df[df["score"] >= 5]
print(f"After removing low scores: {len(df)}")
```

```python
# 5. Remove extra whitespace from title
df["title"] = df["title"].str.strip()
```

Task 3 — Save as CSV

```python
# Save cleaned data
df.to_csv("data/trends_clean.csv", index=False)

print(f"\nSaved {len(df)} rows to 'data/trends_clean.csv' ")
```
