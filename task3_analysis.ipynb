```python
import pandas as pd
import numpy as np
```

Task 1 — Load and Explore 

```python
# Load CSV into DataFrame and Print the first 5 rows
df = pd.read_csv("data/trends_clean.csv")
print("First 5 rows:")
df.head(5)
```

```python
#Shape 
print(f"Loaded data: {df.shape}")
```

```python
# Average score and comments
avg_score = df["score"].mean()
avg_comments = df["num_comments"].mean()

print(f"\nAverage score   : {avg_score:.2f}")
print(f"Average comments: {avg_comments:.2f}")
```

Task 2 — Basic Analysis with NumPy

```python
scores = df["score"].values
comments = df["num_comments"].values

# Mean, Median, Std
print(f"Mean score   : {np.mean(scores):.2f}")
print(f"Median score : {np.median(scores):.2f}")
print(f"Std deviation: {np.std(scores):.2f}")
```

```python
# Max and Min
print(f"Max score    : {np.max(scores)}")
print(f"Min score    : {np.min(scores)}")
```

```python
# Category with most stories
top_category = df["category"].mode()[0]
top_count = (df["category"] == top_category).sum()

print(f"\nMost stories in: {top_category} ({top_count} stories)")
```

```python
# Most commented story
max_comments_idx = np.argmax(comments)
top_story = df.iloc[max_comments_idx]

print(f"\nMost commented story: \"{top_story['title']}\" — {top_story['num_comments']} comments")
```

Task 3 — Add New Columns 

```python
# Engagement formula
df["engagement"] = df["num_comments"] / (df["score"] + 1)

# Popular flag
df["is_popular"] = df["score"] > avg_score
```

```python
# Step 4 — Save Result

df.to_csv("data/trends_analysed.csv", index=False)

print(f"\nSaved to {"data/trends_analysed.csv"}")
```
